📝 Redacción del Problema
Contexto: En la clase de Geometría, los estudiantes necesitan una herramienta rápida y eficaz para verificar sus cálculos sobre figuras planas.

El Reto: Desarrollar un programa interactivo en Python que, mediante un menú de opciones, permita al usuario calcular tanto el área como el perímetro de un rectángulo. El programa debe mantenerse en ejecución hasta que el usuario decida salir, utilizando funciones específicas para cada tarea y estructuras de control para dirigir las decisiones del usuario.

📐 Fórmulas MatemáticasPara un rectángulo con base ($b$) y altura ($h$), utilizaremos las siguientes expresiones:
* Area = b * h
* Perimetro = 2 * (b * h)

![Rectangulo](https://raw.githubusercontent.com/alexandroaguilar128/Act-15-area-perimetro-rectangulo/refs/heads/main/Rectangulo.png)

```python
# ---------------------------------------------------------
# PROGRAMA: Calculadora Geométrica de Rectángulos 📏
# ---------------------------------------------------------

def area_rectangulo():
    print("\n✨ --- Calculando el Área --- ✨")
    base = float(input("🔸 Ingresa la base: "))
    altura = float(input("🔸 Ingresa la altura: "))
    area = base * altura
    print(f"✅ El área resultante es: {area} u²")

def perimetro_rectangulo():
    print("\n✨ --- Calculando el Perímetro --- ✨")
    base = float(input("🔹 Ingresa la base: "))
    altura = float(input("🔹 Ingresa la altura: "))
    perimetro = 2 * (base + altura)
    print(f"✅ El perímetro resultante es: {perimetro} u")

def menu():
    print("\n--- 🛠️  MENÚ DE OPCIONES  ---")
    print("1. 🟦 Calcular área de un rectángulo")
    print("2. 📏 Calcular perímetro de un rectángulo")
    print("3. ❌ Salir")
    opcion = input("👉 Selecciona una opción (1-3): ")
    return opcion

# --- Bloque Principal ---
def ejecutar_programa():
    print("👋 ¡Bienvenido al sistema de geometría interactivo!")

    continuar = True
    while continuar:
        opcion = menu()

        if opcion == "1":
            area_rectangulo()
        elif opcion == "2":
            perimetro_rectangulo()
        elif opcion == "3":
            print("\n🌟 Gracias por usar el programa. ¡Hasta pronto! 🌟")
            continuar = False
        else:
            print("\n⚠️ Opción no válida. Por favor, elige 1, 2 o 3.")

# Llamada inicial para arrancar el código
ejecutar_programa()
```
